```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')   
```

```python
dengue = pd.read_csv("E:\Portable_data\Codes\dengue analysis\dengue.csv")
dengue
```

<h1>1. Dengue Cases Over the Years </h1>

```python


dengue.groupby('Year')['Dengue_Cases'].sum().plot(kind='line', marker='o', figsize=(10,5))
plt.title("Total Dengue Cases by Year")
plt.ylabel("Number of Dengue Cases")
plt.grid(alpha=0.3)
plt.show()
```

<h1>Insights</h1>
<h3>Insight 1: The chart clearly shows that dengue cases peaked in 2019 before dropping sharply in 2020, indicating a possible effect of improved health interventions.
 </h3>
<h3>Insight 2: After 2020, the steady decline suggests successful government campaigns or reduced mosquito activity due to weather patterns.</h3>

<h1>2. Dengue Deaths Over the Years</h1>

```python

dengue.groupby('Year')['Dengue_Deaths'].sum().plot(kind='line', marker='o', color='red', figsize=(10,5))
plt.title("Total Dengue Deaths by Year")
plt.ylabel("Number of Deaths")
plt.grid(alpha=0.3)
plt.show()
```

<h1>Insights</h1>
<h3>Insight 1: The graph reveals that 2016 recorded the highest number of dengue related deaths before significantly dropping in the following year. </h3>
<h3>Insight 2: This decline may reflect improvements in early diagnosis and better access to hospital treatment.</h3>

<h1>3. Average Dengue Cases per Region</h1>

```python

dengue.groupby('Region')['Dengue_Cases'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(10,5))
plt.title("Average Dengue Cases per Region")
plt.ylabel("Average Number of Cases")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

<h1>Insights</h1>
<h3>Insight 1: Regions like NCR and Region IV-A consistently record higher average dengue cases, showing they remain dengue hotspots. </h3>
<h3>Insight 2: On the other hand, regions such as CAR and Region II have notably lower averages, suggesting more effective control measures or lower population density.</h3>

<h1>4. Top 5 Regions with Most Dengue Deaths</h1>

```python

dengue.groupby('Region')['Dengue_Deaths'].sum().sort_values(ascending=False).head(5).plot(kind='bar', color='darkred', figsize=(10,5))
plt.title("Top 5 Regions with Most Dengue Deaths")
plt.ylabel("Total Deaths")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

<h1>Insights</h1>
<h3>Insight 1: The chart highlights that NCR, Region IV-A, and Region III record the highest number of dengue related deaths showing how these densely populated regions face stronger transmission risks.</h3>
<h3>Insight 2: These results imply that urban crowding and poor sanitation likely contribute to the high fatality rate emphasizing the need for intensified local health programs.</h3>

<h1>5. Monthly Dengue Trend Across All Years</h1>

```python

month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
dengue.groupby('Month')['Dengue_Cases'].sum().reindex(month_order).plot(kind='bar', figsize=(12,6))
plt.title("Total Dengue Cases by Month")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

<h1>Insights</h1>
<h3>Insight 1: The bar graph clearly indicates that dengue cases surge between June and September aligning with the rainy season when mosquito breeding increases.
 </h3>
<h3>Insight 2: This pattern shows that preventive campaigns should be intensified before mid year to minimize the seasonal spike in infections. </h3>
